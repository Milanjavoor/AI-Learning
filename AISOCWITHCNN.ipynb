```python
from textwrap import fill
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout,Conv1D,MaxPooling1D,Dense,Flatten
from sklearn.model_selection import train_test_split
import pandas as pd
from google.colab import files
data=files.upload()
df=pd.read_csv("cyber.csv")
df=df.drop(columns=['index', 'timestamp', 'source_ip', 'dest_ip', 'command_line'])
df=df.fillna(method="ffill")
x=df.drop('label',axis=1)
y=df['label']
from sklearn.preprocessing import LabelEncoder
ledic={}
for col in x.select_dtypes(include='object').columns:
  le=LabelEncoder()
  x[col]=le.fit_transform(x[col])
  ledic[col]=le
y=LabelEncoder().fit_transform(y)
from sklearn.preprocessing import MinMaxScaler
xscaled=MinMaxScaler().fit_transform(x)
nfeatures=xscaled.shape[1]
x_cnn=xscaled.reshape(xscaled.shape[0],nfeatures,1)
print("the shape of the data for the cnn model is ",x_cnn.shape)
x_train,x_test,y_train,y_test=train_test_split(x_cnn,y,test_size=0.2,random_state=42)
model=Sequential()
model.add(Conv1D(filters=32,kernel_size=3,input_shape=(x_cnn.shape[1],1),activation="relu"))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
history=model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),batch_size=32)
loss,accuracy=model.evaluate(x_test,y_test)
print("accuracy score=",accuracy)
def predict_event(model, scaler, input_data):
    import numpy as np
    input_data = np.array(input_data).reshape(1, -1)
    input_scaled = scaler.transform(input_data)
    input_cnn = input_scaled.reshape(1, input_scaled.shape[1], 1)
    prediction = model.predict(input_cnn)
    if prediction[0][0] > 0.5:
        return "🚨 Attack Detected"
    else:
        return "✅ Normal Traffic"






```
